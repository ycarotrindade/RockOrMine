# Import

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split, StratifiedKFold, RandomizedSearchCV
from sklearn.preprocessing import LabelEncoder
import warnings
import pickle
```

```python
DATABASE_PATH = ""
df = pd.read_csv(DATABASE_PATH)
```

```python
X = df.iloc[:,:60].values
y = df['R'].values
```

```python
encoder = LabelEncoder()
y = encoder.fit_transform(y)
```

```python
warnings.filterwarnings('ignore')
params = {
    'n_estimators':list(range(100,300,15)),
    'criterion':['gini','entropy','log_loss'],
    'min_samples_split':list(range(1,7)),
    'max_depth':list(range(100,300,15))
}
classifier_test = RandomForestClassifier(random_state=5)
cv = StratifiedKFold(n_splits=10,shuffle=True,random_state=5)
randomized = RandomizedSearchCV(
    estimator = classifier_test,
    param_distributions=params,
    cv=cv,
    random_state=5
)
randomized.fit(X,y)
print(randomized.best_score_)
print(randomized.best_params_)
best_params = randomized.best_params_
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,shuffle=True,random_state=5)
```

```python
classifier = RandomForestClassifier(random_state=5,**best_params)
classifier.fit(X_train,y_train)
preds = classifier.predict(X_test)
print(accuracy_score(y_test,preds))
```

```python
response  = input("Do you want to save the model? [Y/n]: ").lower()
if response == 'y':
  with open('model.pkl','wb') as f:
    pickle.dump(classifier,f)
```
